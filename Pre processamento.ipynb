```python
# função astype transforma o tipo de dado para object, int, float, etc
dados['DataYear'] = dados['DataYear'].astype(int)   
```

```python
# Dados missing
# dados.dropna() -> exclui os registros com dados missing
# dados.isnull() -> retorna True se for dado NaN
dados['ENERGYSTARScore'].fillna(dados['ENERGYSTARScore'].mean(), inplace = True)
```
