```python
import pandas as pd
pd.set_option('display.max_columns', 23)
dados = pd.read_csv('C:\\Python\\Didatica Tech\\Datasets\\recipeData.csv', encoding='latin-1') # 
dados.drop('URL', axis = 1, inplace = True)
```

```python
# carrega apenas registros em que o valor da coluna StyleID seja igual aos IDs da lista
# neste exercício foram eliminados itens com menos de 1000 registros, para manter apenas os mais relevantes
dados = dados.loc[(dados['StyleID'].isin([7, 10, 134, 9, 4,30, 86, 12,92, 6, 175, 39]))]
```

```python
# removendo colunas com mais de 50% de registros com valor nulo
dados.drop('PrimingMethod', axis = 1, inplace = True)
dados.drop('PrimingAmount', axis = 1, inplace = True)
dados.drop('UserId', axis = 1, inplace = True)
dados.drop('BeerID', axis = 1, inplace = True)
dados.drop('Name', axis = 1, inplace = True)
dados.drop('Style', axis = 1, inplace = True)
```

```python
faltantes = dados.isnull().sum()
faltantes_percentual = (dados.isnull().sum() / len(dados['StyleID'])) * 100
print(faltantes_percentual)
```

```python
dados.head()
```

```python
# Transformando os dados em 0 e 1
#dados["SugarScale"].replace(to_replace=dict('Specific Gravity'=1, Plato=0), inplace=True)
dados["SugarScale"].replace('Specific Gravity', 1, inplace=True)
dados["SugarScale"].replace('Plato', 0, inplace=True)
```

```python
class_encode = pd.get_dummies(dados['BrewMethod'])  
# concatena as novas colunas do método encoding com o dataset
dados = pd.concat([dados, class_encode], axis=1)
dados.drop('BrewMethod', axis = 1, inplace = True)
```

```python
dados['BoilGravity'].fillna(dados['BoilGravity'].median(), inplace=True)
```

```python
# o objetivo é preencher os valores nulos com o valor estimado por algoritmo
# a variável com campos nulos será a variável target
# 1) carrega todas as colunas cujo valor da variável PitchRate seja não nulos
x_treino = dados[dados['PitchRate'].notnull()] 
# 2) apaga a colua PitchRate porque será a variável target
x_treino.drop('PitchRate', axis = 1, inplace = True)
# 3) carrega apenas a coluna 'PitchRate' com os valores não nulos
y_treino = dados[dados['PitchRate'].notnull()]['PitchRate']

# 1) carrega todas as colunas cujo valor da variável PitchRate seja nulo para ser preenchido
x_preench = dados[dados['PitchRate'].isnull()]
# 2) carrega apenas a coluna 'PitchRate' com os valores nulos
y_preench = dados[dados['PitchRate'].isnull()]['PitchRate']
# 3) apaga a colua PitchRate porque será a variável target
x_preench.drop('PitchRate', axis = 1, inplace = True)
```

```python
x_preench
```

```python
# excluir as variáveis que possuem valor nulo
x_treino.drop('MashThickness', axis = 1, inplace = True)
x_treino.drop('PrimaryTemp', axis = 1, inplace = True)
x_preench.drop('MashThickness', axis = 1, inplace = True)
x_preench.drop('PrimaryTemp', axis = 1, inplace = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
modelo = DecisionTreeRegressor()
modelo.fit(x_treino, y_treino)
#predição dos valores da variável
y_preench = modelo.predict(x_preench)
```

```python
#substituir os valores nulos pelos previstos
dados.PitchRate[dados.PitchRate.isnull()] = y_preench
```

```python
faltantes = dados.isnull().sum()
faltantes_percentual = (dados.isnull().sum() / len(dados['StyleID'])) * 100
print(faltantes_percentual)
```
