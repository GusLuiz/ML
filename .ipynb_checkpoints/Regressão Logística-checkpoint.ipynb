```python
import pandas as pd
pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows', 64)
arquivo = pd.read_csv('C:\Python\Didatica Tech\Datasets\Data_train_reduced.csv')
```

```python
arquivo.drop('Product', axis = 1, inplace=True)
```

```python
faltantes = arquivo.isnull().sum()
faltantes_percentual = (faltantes / len(arquivo['Product.ID'])) * 100
print(faltantes_percentual)
```

```python
arquivo.drop('Respondent.ID', axis = 1, inplace=True) # info não relevante
# drop em colunas que tenham mais de 20% de dados faltantes 
arquivo.drop('q8.2', axis = 1, inplace=True)
arquivo.drop('q8.8', axis = 1, inplace=True)
arquivo.drop('q8.9', axis = 1, inplace=True)
arquivo.drop('q8.10', axis = 1, inplace=True)
arquivo.drop('q8.17', axis = 1, inplace=True)
arquivo.drop('q8.18', axis = 1, inplace=True)
arquivo.drop('q8.20', axis = 1, inplace=True)
# variável que resolveria o problema por ter correlação direta com a variável target. Excluída para não prejudicar o exercício
arquivo.drop('q1_1.personal.opinion.of.this.Deodorant', axis = 1, inplace=True) 
```

```python
arquivo['q8.12'].fillna(arquivo['q8.12'].median(), inplace=True)
arquivo['q8.7'].fillna(arquivo['q8.7'].median(), inplace=True)
```

```python
y = arquivo['Instant.Liking'] # Variável target
x = arquivo.drop('Instant.Liking', axis = 1) # variáveis preditoras
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

stratifiedkfold = StratifiedKFold(n_splits = 5)

modelo = LogisticRegression()
resultado = cross_val_score(modelo, x, y, cv = stratifiedkfold)

print(resultado.mean())
```

```python
# Ajustando parâmetros para buscar a melhor acurácia
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

Valores_C = np.array([0.01, 0.1,0.5,1,2,3,5,10,20,50,100])
regularizacao = ['l1','l2']
valores_grid = {'C':Valores_C, 'penalty':regularizacao}

modelo = LogisticRegression()

grid_regressao_logistica = GridSearchCV(estimator = modelo, param_grid = valores_grid, cv=5)
grid_regressao_logistica.fit(x,y)

print("Melhor acurácia: ", grid_regressao_logistica.best_score_)
print("Parâmetro C: ", grid_regressao_logistica.best_estimator_.C)
print("Regularização: ", grid_regressao_logistica.best_estimator_.penalty)
```

```python
# Exercício
from sklearn.datasets import load_breast_cancer

pd.set_option('display.max_columns', 30)
dados = load_breast_cancer() # está em formato específico do sklearn, precisa ser transformado em dataframe pandas
xe = pd.DataFrame(dados.data, columns=[dados.feature_names])
ye = pd.Series(dados.target)
```

```python
xe.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

stratifiedkfold = StratifiedKFold(n_splits = 5)

modelo = LogisticRegression()
resultado = cross_val_score(modelo, xe, ye, cv = stratifiedkfold)

print(resultado.mean())
```

```python
# Ajustando parâmetros para buscar a melhor acurácia
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

Valores_C = np.array([0.01, 0.1, 10, 20, 50, 60, 70, 80, 90, 100, 110]) # valores podem ser ajustados para buscar a melhor acurácia do modelo
#regularizacao = ['l1','l2']
valores_grid = {'C':Valores_C} # , 'penalty':regularizacao

modelo = LogisticRegression(penalty = 'l1')

grid_regressao_logistica = GridSearchCV(estimator = modelo, param_grid = valores_grid, cv=5)
grid_regressao_logistica.fit(xe,ye)

print("Melhor acurácia: ", grid_regressao_logistica.best_score_)
print("Parâmetro C: ", grid_regressao_logistica.best_estimator_.C)
print("Regularização: ", grid_regressao_logistica.best_estimator_.penalty)
```
